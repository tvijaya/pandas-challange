```python
import pandas as pd
import json
```

```python
json_obj = json.loads(open('purchase_data2.json').read())
json_obj
```

```python
purchase_df = pd.DataFrame(json_obj)
purchase_df.head()
```

```python
columns = purchase_df.columns.values
columns
```

```python
Total_Players = purchase_df['SN'].count()
Total_Players
```

```python
unique_items_count = purchase_df['Item ID'].nunique()
unique_items_count
```

```python
average_price = round((purchase_df['Price'].sum())/(purchase_df['Item ID'].count()),2)
average_price
```

```python
total_purchases = purchase_df['Price'].count()
total_purchases
```

```python
total_revenue = round(purchase_df['Price'].sum(),2)
total_revenue
```

```python
purchasing_analysis = pd.DataFrame({'Number of Unique Items' : unique_items_count,
                                    'Average Purchase Price' : average_price,
                                    'Total Number of Purchases': total_purchases,
                                    'Total Revenue': total_revenue
                                   }, index=[0])
purchasing_analysis
```
